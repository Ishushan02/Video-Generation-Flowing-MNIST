```python
from torchvision.datasets import MovingMNIST
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image as IPyImage, display
```

```python
transform = transforms.ToTensor()

dataset = MovingMNIST(
    root='./data/',
    split='train',
    split_ratio=10,
    download=True,
    transform=transform
)
```

```python
data = np.load("data/MovingMNIST/mnist_test_seq.npy")
data.shape
```

```python
data.shape
```

```python
sequence = data[:, 5, :, :]

frames = [Image.fromarray((frame * 255).astype(np.uint8), mode='L') for frame in sequence]

gif_path = 'datavisualization/moving_mnist_5.gif'
frames[0].save(
    gif_path,
    save_all=True,
    append_images=frames[1:],
    duration=100,  # milliseconds between frames
    loop=0  # loop forever
)

print(f"GIF saved to {gif_path}")
display(IPyImage(filename=gif_path))
```
